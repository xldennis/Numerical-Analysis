In numerical analysis, Richardson extrapolation is a sequence acceleration method, used to improve the rate of convergence of a sequence.

$$f^{(2)}(x) \approx \frac{f(x+h)-2f(x)+f(x-h)}{h^2}$$

the truncation error is $O(h^2)$

Taylor Series of f(x) about the point x and evaluated at x + h and x âˆ’ h leads to the central difference formula:

$$f'(x) = \frac{f(x+h)-(x+h)}{2h} - \frac{h^2}{6}f^{(3)}(x) - \frac{h^4}{120}f^{(5)}(x) - \cdots$$

Rewrite this in the following form:

$$f'(x_0) = N(h) - \frac{h^2}{6}f^{(3)}(x_0) - \frac{h^4}{120}f^{(5)}(x_0) - \cdots$$

where $N(h) = \frac{f(x_0+h)-(x_0+h)}{2h}$ and replace h by $\frac{h}{2}$, we have

$$f'(x_0) = N(\frac{h}{2}) - \frac{h^2}{4 \times 6}f^{(3)}(x_0) - \frac{h^4}{120 \times 16}f^{(5)}(x_0) - \cdots$$

Careful substraction cancels a higher order term and we will have:

$$3f'(x_0) = 4N(\frac{h}{2}) - N(h) + \frac{h^4}{160}f^{(5)}(x_0) + \cdots$$

Where the error is a degree of $O(h^4)$

Use a Richardson Extrapolation to estimate the following limit to 5 decimal places:  

$$ \lim_{n \rightarrow \infty} \left( 1+\frac{d}{n}\right)^{n/c} $$

(That is, until approximations agree to 5 places after the decimal.)  

```python
# E00404624
a = 2; b = 4; c = 4; d = 6
from __future__ import division
```

As $n \to \infty$, $f(x) = f(x+1) \Rightarrow f(x) = (x+1)f(x+1) - xf(x) \tag{1}$

And 
$$
\begin{align*}
(2x+1)f(x) &= (x+1)^2f(x+1) - (x)^2f(x) \tag{2}\\
(2(x+1)+1)f(x+1) &= (x+2)^2f(x+2) - (x+1)^2f(x+1) \tag{3}
\end{align*}$$

(3) - (2) and since $f(x) = f(x+1) \ when \ n \to \infty$, we get:

$$f(x) = \frac{(x+2)^2f(x+2) + (x)^2f(x)- 2(x+1)^2f(x+1)}{2} \tag{4}$$

```python
def f(n):
    return (1+d/n)**(n/c)

tolerance = 10**(-5) #5 digit accuracy

def Richardson_1(n): #equation(1)
    return (n+1)*f(n+1)-n*f(n)

def Richardson_2(n): #equation(4)
    return (n**2*f(n)-2*(n+1)**2*f(n+1)+(n+2)**2*f(n+2))/2

fn = f(1)
print('Equation (1) Richarson Extrapolation')
print('{:^0s} {:^13s} {:>12s} {:>9s}'.format('n','n-1','|f(n)-f(n-1)|','f(n)' ) ) 

for n in range(2,1000):
    fp = fn
    fn = Richardson_1(n)
    
    print('{:0.0f} {:6.0f} {:15.6f} {:22.10f}'.format(n,n-1,abs(fn-fp), fn )  ) 
    
    if( abs(fn-fp) < tolerance):
        break
```

```python
fn = f(1)
print('Equation (2) Richarson Extrapolation')
print('{:^0s} {:^13s} {:>12s} {:>9s}'.format('n','n-1','|f(n)-f(n-1)|','f(n)' ) ) 

for n in range(2,1000):
    fp = fn
    fn = Richardson_2(n)
    
    print('{:0.0f} {:6.0f} {:15.6f} {:22.10f}'.format(n,n-1,abs(fn-fp), fn )  ) 
    
    if( abs(fn-fp) < tolerance):
        break
```

Write a procedure as a function of $h$ that applies the trapezoidal rule for definite integrals to  

$$ \int_0^1 \ln\left(1+x^{d/c}\right) dx $$

Then apply Romberg integration until 5 decimal places of accuracy are obtained.  
   

$$\int_a^b f(x)\,dx \approx  \frac{h}{2} \left\{   f(x_0)+ 2f(x_1)+2f(x_2)+2f(x_3)+\dots +2f(x_{n-1}) + f(x_n) \right\}\tag{1}$$

```python
def f(x):
    log(1+x**(d/c))

def trapezoid(a,b,h):
    n = int((b-a)/h)
    x0 = a
    output =f(a)
    for i in range(2,n+1):
        x  = x + h
        output = output + 2. * f(x)
    output = (output + f(b)) * h/2
    return output
```

$
\begin{align*}
R(0,0) &= \frac{1}{2} (b-a) (f(a) + f(b)) \\
R(n,0) &= \frac{1}{2} R(n-1,0) + h_n \sum_{k=1}^{2^{n-1}} f(a + (2k-1)h_n) \\
R(n,m) &= R(n,m-1) + \frac{1}{4^{m}-1} (R(n,m-1) - R(n-1,m-1))
\end{align*}
$

where $h_n = \frac{b-a}{2^n}$

```python
def romberg(f, a, b, eps=1e-5):
    R = [[(b - a)/2 * (f(a) + f(b))]]  # R[0][0]
    n = 1
    while True:
        h = float(b - a) / 2 ** n
        R.append([None] * (n + 1))  # Add an empty row.
        R[n][0] = R[n-1][0]/2 + h*sum(f(a+(2*k-1)*h) for k in xrange(1, 2**(n-1)+1))
        for m in xrange(1, n+1):
            R[n][m] = R[n][m-1] + (R[n][m-1] - R[n-1][m-1]) / (4 ** m - 1)
        if abs(R[n][n-1] - R[n][n]) < eps:
            return R[n][n]
        n += 1
```

b) (5 points) The partial sums for an alternating series at best are of the form 

$$ S(n) = L + (-1)^{n} \frac{A}{n} + (-1)^{n} \frac{B}{n^2} + \ldots $$

Using $S(n+1)$ and $S(n)$ does not work.  Instead, notice that $(-1)^{n+2} = (-1)^n$, so that

\begin{eqnarray}
(n+2) \cdot S(n+2) & = & (n+2)\cdot L + (-1)^{n} A+ (-1)^{n} \frac{B}{n+2} + \ldots \\
n \cdot S(n) & = & \phantom{92+9}n \cdot L + (-1)^{n} A+ (-1)^{n} \frac{B}{n} + \ldots
\end{eqnarray}

from which we get 

$$ \frac{(n+2)S(n+2) - nS(n)}{2} = L + (-1)^n\frac{B}{n^2+2n} + \ldots $$

Consequently, the approximation alternates about $L$, so that a good approximation for $L$ is the average of successive approximations, or 

$$ \frac{1}{2} \cdot \left( \frac{(n+2)S(n+2) - nS(n)}{2} + \frac{(n+3)S(n+3) - (n+1)S(n+1)}{2} \right) $$
