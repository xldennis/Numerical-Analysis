```python
lines = [line.rstrip('\n') for line in open('1.txt')]
lines.pop(1);lines.pop(1)

data = []
theta = []
omega = []
time = []

for i in range(1,len(lines)):
    data = lines[i].split()
    time.append(data[1])
    theta.append(data[3])
    omega.append(data[4])
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots()
ax1.plot(time,theta,label="theta")
ax1.set_xlabel('time (s)')
ax1.set_ylabel('theta (deg)', color='b')


ax2 = ax1.twinx()
ax2.plot(time,omega,label="omega",color='g')
ax2.set_ylabel('omega (rad/s)', color='g')

plt.title("Pendulum motion for l = 0.5 m and theta_0 = 1 deg",y=-0.3)
ax1.legend(loc = (0., 1.02), frameon = False)
ax2.legend( loc = (0.3, 1.02), frameon = False)

plt.show()
```

the period of theta and omega are almost the same. Theta varies between $1^o$ and $-1.4^o$ and omega varies between 0.9 rad/s and -0.9 rad/s
